### Calcula la más corta de las FCS de $\sqrt{n_1}$ y $\sqrt{n_2}$.

Comprobamos que $n_1$ y $n_2$ no sean cuadrados perfectos

```julia
sqrt(n1)
```

```julia
sqrt(n2)
```

Calculamos la FCS para el caso $\alpha = \sqrt{d}$. Gracias a esta restricción podemos usar $Q_i = 1$ como condición de parada.

```julia
# Función que calcula la FCS de un número de la forma √d
function fcs(d)
    P = 0
    Q = 1
    fsd = floor(Int, sqrt(d))
    q = fsd
    i = 0
    
    println("Paso | P_i | Q_i | q_i")
    println("$i | $P | $Q | $q")
    
    i = 1
    P = q*Q - P
    Q_prev = Q
    Q = div(d - P^2, Q)
    q = div(P + fsd, Q)
    
    println("$i | $P | $Q | $q")
    
    while true
        i += 1
        P_prev = P
        P = q*Q - P
        Q_prev2 = Q_prev
        Q_prev = Q
        Q = Q_prev2 + q*(P_prev - P)
        q = div(P + fsd, Q)
        
        println("$i | $P | $Q | $q")
        Q_prev == 1 && break
    end
end
```

Aplicando esta función a $n_1$ y a $n_2$ tenemos

```julia
fcs(n1)
```

```julia
fcs(n2)
```

Por tanto $\sqrt{n_1} = [807, \overline{1, 2, 1, 32, 4, ..., 1, 1614}]$ y $\sqrt{n_2} = [229, \overline{28, 1, 1, 1, 2, ..., 28, 458}]$.
