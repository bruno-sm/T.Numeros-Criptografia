# Apartado 2

### Calcula $a^{n-1}\ mod\ n$ para cada una de esas cuatro bases usando sucesivamente el algoritmo de izda-drcha y de drcha-izda.

Primero programamos ambos algoritmos



### Izquierda-derecha

```julia
using Base.Iterators

function izquierda_derecha(a, n)
    # Primero calculamos n-1 en base 2
    binary_digits = []
    b = n - 1
    while b != 1
        push!(binary_digits, b % 2)
        b = div(b, 2)
    end
    push!(binary_digits, 1)

    acu = 1
    exp = 0

    println("  Paso  |  acu  ")	
    println("----------------")
    println("0 | $acu ")

    for (i, e_i) in enumerate(reverse(binary_digits))
        acu ^= 2 
        exp *= 2

        if e_i == 1
            acu *= a
            exp += 1
        end

        acu %= n
        
        println("$i | $acu ")
    end

    return acu
end;
```

### Derecha-izquierda

```julia
function derecha_izquierda(a, n)
    base = a
    acu = 1
    num = n-1 
    exp = 0

    println("  Paso  |  acu  |  base  ")	
    println("-------------------------")
    println("0 | $acu | $base")
    i = 1 
    while num > 0
        if num % 2 != 0
            acu = (acu * base) % n
            num = (num-1)/2
            e_i = 1
        else
            num = num/2
            e_i = 0
        end
        base = (base^2) % n
        exp = exp + e_i*2^i

        println("$i | $acu | $base")
        i = i+1
    end

    return acu
end;
```

Aplicamos estas funciones a cada una de las bases:

```julia
for a in [2, 3, 5, 7, 11]
    println("Izquierda-derecha en $a^$(n-1) mod $n\n")
    izquierda_derecha(a, n)
    println("\n\nDerecha-izquierda en $a^$(n-1) mod $n\n")
    derecha_izquierda(a, n)
    println("\n\n")
end
```

### \textquestiondown Es n un posible primo de Fermat para alguna de ellas? \textquestiondown Es n un pseudoprimo para alguna de ellas?

De la ejecución anterior podemos ver que $2^{n-1} \equiv_n 1$, $3^{n-1} \equiv_n 1$, $5^{n-1} \equiv_n 1$, $7^{n-1} \equiv_n 1$ y $11^{n-1} \equiv_n 1$, por tanto es un posible primo de Fermat para todas ellas. Dado que el test TPF con estas bases no nos ha podido asegurar que n no sea primo, para comprobar si es pseudoprimo lo descompondremos.

```julia
using Primes
factor(n)
```

Vemos así que $71304917$ realmente es un primo.
