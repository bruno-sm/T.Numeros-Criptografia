### Factoriza n-1 aplicando el método de Fermat o el $\rho$ de Polard.

**Método de Fermat:**

```julia
function fermat(n)
    s = ceil(Int, sqrt(n))
    
    if n == (s-1)^2
        return s - 1, s - 1
    end
    
    println("Paso | r | u | v")
    
    r = s^2 - n
    u = 2*s + 1
    v = 1
    i = 0
    
    println("$i | $r | $u | $v")
        
    while r != 0
        i += 1
        if r > 0
            r = r - v
            v = v + 2
        else
            r = r + u
            u = u + 2
        end
        
        println("$i | $r | $u | $v")
    end
    
    return Int((u-v)/2), Int((u+v-2)/2)
end;
```

Comprobamos que el algoritmo funciona como en el ejemplo

```julia
fermat(391)
```

Sin embargo para $n-1$ parece que necesita demasiadas iteraciones.

```julia
fermat(n-1)
```

**$\rho$ de Polard con detección de Floyd:**

```julia
function polard(n, t = 100, f = x -> x^2 + 1)
    x = 1
    y = x
    i = 0
    
    println("Paso | x | y | mcd")
    println("$i | $x | $y | -")
    
    while i < t
        i += 1
        x = f(x) % n
        y = f(f(y)) % n
        g = gcd(x - y, n)
        
        println("$i | $x | $y | $g")
        
        if 1 < g && g < n
            return g, Int(n/g)
        end
    end
    
    error("No hay divisores con $t iteraciones.")
end
```

Vemos que la $\rho$ de Polard sí es capaz de factorizar rápidamente $n-1$

```julia
polard(n-1)
```

Para factorizar completamente un número construimos una función que, independientemente del método utilizado, descompone cada factor de forma recursiva.

```julia
using Primes

function factorizacion(n, metodo)
    println("\n\nFactorizando $n:\n")
    d1, d2 = metodo(n)
    println("\n$n = $d1 * $d2")
    
    if !isprime(d1)
        d1 = factorizacion(d1, metodo)
    end
    
    if !isprime(d2)
        d2 = factorizacion(d2, metodo)
    end
    
    return d1, d2
end;
```

Factorizamos completamente $n-1$ usando la $\rho$ de Polard.

```julia
factorizacion(n - 1, polard)
```

### Aplica el teorema de Lucas-Lehmer para encontrar un certificado de primalidad de tu número.

Gracias al apartado anterior conocemos los factores primos de $n-1$

```julia
fp = [2, 13, 23, 43, 47, 59];
```

Buscamos entonces un elemento primitivo para $n$

```julia
for a in 2:n-1
    primitivo = powermod(a, n-1, n) == 1
    for p in fp
        primitivo &= powermod(a, Int((n-1)/p), n) != 1
    end
    
    if primitivo
        println("$a es un elemento primitivo")
        break
    end
end
```

Por lo tanto $2$ es un certificado de la primalidad de $n$.
